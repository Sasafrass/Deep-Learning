```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Path for csv files
path = "csv/"

# Function that takes as input strings and outputs the corresponding plot
def plotter(units, steps, lr, dropout, ylim = (0,200), CNN = False, nump = False):
    
    # Check for CNN
    if CNN:
        file = "csv/CNN.csv"
        df = pd.read_csv(file, header = None)
        
    # Check for numpy
    if nump:
        file = "csv/nump.csv"
        df = pd.read_csv(file, header = None)
    
    else:
        # Build file string using the input strings
        file = path + str(units) + "-" + str(steps) + "-" + str(lr) + "-" + str(dropout)
        df = pd.read_csv(file + ".csv", header = None)
        
    print(df)
    
    # Get losses and accuracies
    trainloss = df.iloc[0,:]
    trainacc  = df.iloc[1,:]
    testloss  = df.iloc[2,:]
    testacc = df.iloc[3,:] 
    steps   = df.iloc[4,:]
    
    # Create subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12,4))
    
    # Everything for the accuracy plot
    ax1.plot(steps, trainacc, label = "Train")
    ax1.plot(steps, testacc, label = "Test")
    ax1.legend()
    ax1.set_title("Train and Test Accuracy")
    
    # Everything for the loss plot
    ax2.plot(steps, trainloss, label = "Train")
    ax2.plot(steps, testloss, label = "Test")
    ax2.set_ylim(ylim)
    ax2.legend()
    ax2.set_title("Train and Test Loss")

    # Save figure
    plt.savefig(file + '.png', format='png', bbox_inches = 'tight', pad_inches = 0)
    
```

# PyTorch MLP

# Default Parameters

```python
plotter(100, 1500, 0.002, False, ylim = (0,5))
```

# More steps

```python
plotter(100, 3000, 0.002, False, ylim = (0,5))
```

# More Layers

```python
plotter("(200,100,50)", 1500, 0.002, False, ylim = (0,5))
```

# Even More Layers

```python
plotter("(800,400,200,100,50)", 1500, 0.002, False, ylim = (0,5))
```

# Higher Learning Rate

```python
plotter("(800,400,200,100,50)", 1500, 0.005, False, ylim = (0,5))
```

# First time Dropout

```python
plotter("(2000,1000,500,250,100)", 3000, 0.002, True, ylim = (0,5))
```

# Deeper Network with no dropout

```python
plotter("(1600,800,400,200,100,50,25)", 3000, 0.002, False, ylim = (0,5))
```

# Deeper Network, dropout 0.2 only after first linear

```python
plotter("(1600,800,400,200,100,50,25)", 3000, 0.002, True, ylim = (0,5))
```

# PyTorch CNN

# NumPy MLP

```python
plotter("", "", "", "nump", ylim = (0,3), nump = True)
```
